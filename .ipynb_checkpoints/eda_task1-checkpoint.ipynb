```python
import pandas as pd
df = pd.read_csv('insurance.csv')
df.head()
```

```python
numerical = ['age', 'bmi', 'children', 'charges']  # Proxy for TotalPremium, TotalClaims
print(df[numerical].describe())
```

```python
print(df.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df['charges'].hist(bins=20)
plt.title('Distribution of Charges (Claims Proxy)')
plt.show()
sns.countplot(x='region', data=df)
plt.title('Distribution of Region (Province Proxy)')
plt.show()
```

```python
corr = df[numerical].corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation Matrix')
plt.show()
sns.scatterplot(x='bmi', y='charges', hue='smoker', data=df)
plt.title('BMI vs Charges by Smoker (Risk Proxy)')
plt.show()
```

```python
sns.boxplot(x='region', y='charges', data=df)
plt.title('Charges by Region')
plt.show()
```

```python
sns.boxplot(x=df['charges'])
plt.title('Box Plot for Charges Outliers')
plt.show()
q1 = df['charges'].quantile(0.25)
q3 = df['charges'].quantile(0.75)
iqr = q3 - q1
outliers = df[(df['charges'] < q1 - 1.5*iqr) | (df['charges'] > q3 + 1.5*iqr)]
print('Outliers count:', outliers.shape[0])
```

```python
# Plot 1: Violin for Charges by Region
sns.violinplot(x='region', y='charges', data=df)
plt.title('Charges Distribution by Region')
plt.show()
```

# Plot 2: Line for Average Charges by Age
avg = df.groupby('age')['charges'].mean().reset_index()
sns.lineplot(x='age', y='charges', data=avg)
plt.title('Temporal Trend: Charges by Age (Proxy for Time)')
plt.show()

```python
# Plot 3: Heatmap for Charges by Smoker and Sex
pivot = df.pivot_table(values='charges', index='sex', columns='smoker', aggfunc='mean')
sns.heatmap(pivot, annot=True)
plt.title('Average Charges by Gender and Smoker')
plt.show()
```

Overall Loss Ratio: Proxy mean charges 13270, assume premium 1.2x = loss ~0.83. Varies by region (southeast high), smoker (yes high), sex (male higher).
Distributions: Charges skewed (skew 1.52), outliers 139.
Temporal Trends: Charges increase with age (proxy for time).
Vehicle Makes: Proxy smoker 'yes' high, 'no' low.
